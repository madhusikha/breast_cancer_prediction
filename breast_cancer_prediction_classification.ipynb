```python
# lets separate input features and target columns

X = df.drop('diagnosis',axis=1)
y = df['diagnosis']
```

```python
X.shape
```

```python
# train test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#  there are no null values in the dataset now

# lets scale the data using MinMax scaler

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

# 1. logistic regression- baseline model

```python

from sklearn.linear_model import LogisticRegression

logit_reg_baseline_model = LogisticRegression()

logit_reg_baseline_model.fit(scaled_X_train, y_train)

y_pred_logit_reg_blm = logit_reg_baseline_model.predict(scaled_X_test)   # blm- baselinemodel
```

```python
from sklearn.metrics import r2_score, accuracy_score
```

```python
r2_logit_blm = r2_score(y_test, y_pred_logit_reg_blm)
r2_logit_blm
```

```python
acc_score_logit_blm = accuracy_score(y_test, y_pred_logit_reg_blm)
acc_score_logit_blm
```

```python
# function to calculate performance metrics
def perf_metrics(y_pred):
    r2 = r2_score(y_test, y_pred)
    acc_score = accuracy_score(y_test, y_pred)
    return r2, acc_score
```

with **simple logistic regression** and **using all the features**, achieved an **R2 score of 0.85** and **accuracy score of 0.965**.

# PCA with Logistic Regression

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.90)
```

```python
# transforming training and testing data using PCA

X_train_pca = pca.fit_transform(scaled_X_train)

X_test_pca = pca.transform(scaled_X_test)
```

```python
X_train_pca.shape

# 30 features in input X are brought down to 6 features after performing PCA
```

# now fitting Logistic Regression on the PCA data

# 3. logistic regression with PCA data- baseline model

```python
from sklearn.linear_model import LogisticRegression

pca_logit_reg_base = LogisticRegression()

pca_logit_reg_base.fit(X_train_pca, y_train)

y_pred_pca_logit_reg = pca_logit_reg_base.predict(X_test_pca)   # blm- baselinemodel

r2_pca_logit_base, acc_score_pca_logit_base = perf_metrics(y_pred_pca_logit_reg)

print(r2_pca_logit_base, acc_score_pca_logit_base)
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

Observations from Logistic Regression with PCA data:

pca = PCA(n_components=0.90)

This means we are informing to select number of features such that the total variance is 90%

```python
# lets plot % variance vs number of components and r2 score and accuracy
```

```python
n_components = np.arange(0.8,1,0.01)
n_components
```

```python
n_components = np.arange(0.8,1,0.01)
r2=[]
acc=[]
number_components = []
for n in range(len(n_components)):
    pca = PCA(n_components=n_components[n])
    X_train_pca = pca.fit_transform(scaled_X_train)
    X_test_pca = pca.transform(scaled_X_test)
    pca_logit_reg_base = LogisticRegression()

    pca_logit_reg_base.fit(X_train_pca, y_train)

    y_pred_pca_logit_reg = pca_logit_reg_base.predict(X_test_pca)   # blm- baselinemodel

    r2_pca_logit_base, acc_score_pca_logit_base = perf_metrics(y_pred_pca_logit_reg)
    number_components.append(pca.n_components_)
    r2.append(r2_pca_logit_base)
    acc.append(acc_score_pca_logit_base)

    print(f'%variance: {round(n_components[n]*100,2)}, Number of components: {pca.n_components_}, R2 score: {round(r2_pca_logit_base,4)}, Accuracy: {round(acc_score_pca_logit_base,4)}')
```

From the above results, we can observe that the maximum R2 score is 0.8743 and Accuracy is 0.9708 for number of components of 5 and 6 (i.e., % variance of 84% to 90%). Beyond which the performance again drops slightly. 

These metrics values are coinciding with Logistic Regression GridSearchCV values (Why?).

```python
plt.plot(number_components,acc)
plt.plot(number_components,r2)
plt.plot(number_components,n_components)
```
